# EDA

```python
# List of important categorical features to analyze
categorical_features = [
    'Gender', 'Air Pollution', 'Alcohol use', 'Dust Allergy', 
    'OccuPational Hazards', 'Genetic Risk', 'chronic Lung Disease', 
    'Balanced Diet', 'Obesity', 'Smoking', 'Passive Smoker', 'Snoring'
]

for feature in categorical_features:
    plt.figure(figsize=(12, 7))
    sns.countplot(x=feature, hue='Level', data=df, palette='inferno')
    plt.title(f'Distribution of Disease Level by "{feature}"', fontsize=16)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Number of Patients', fontsize=12)
    plt.legend(title='Disease Level')
    plt.xticks(rotation=45) 
    filename = f"plots/{feature.replace(' ', '_').lower()}_distribution.png"
    plt.savefig(filename, bbox_inches='tight')
    plt.show()
```

### Visualize Alcohol use vs. Cancer Level

```python
# Visualize Alcohol use vs. Cancer Level
sns.catplot(data=df, x='Alcohol use', col='Level', kind='count', 
            col_order=['Low', 'Medium', 'High'], palette='viridis', height=5, aspect=0.8)
plt.suptitle('Patient Count by Alcohol Use Across Cancer Levels', y=1.03, fontsize=16)
plt.show()
```

### Visualize Air Pollution vs. Cancer Level

```python
# Visualize Air Pollution vs. Cancer Level
sns.catplot(data=df, x='Air Pollution', col='Level', kind='count', 
            col_order=['Low', 'Medium', 'High'], palette='plasma', height=5, aspect=0.8)
plt.suptitle('Patient Count by Air Pollution Across Cancer Levels', y=1.03, fontsize=16)
plt.savefig("plots/air_pollution_vs_cancer_level.png", bbox_inches='tight')
plt.show()
```

### Visualize Genetic Risk vs. Cancer Level

```python
# Visualize Genetic Risk vs. Cancer Level
plt.figure(figsize=(12, 7))
sns.boxplot(data=df, x='Level', y='Genetic Risk', order=['Low', 'Medium', 'High'], palette='crest')
plt.title('Distribution of Genetic Risk by Cancer Level', fontsize=16)
plt.xlabel('Cancer Level', fontsize=12)
plt.ylabel('Genetic Risk Intensity', fontsize=12)
plt.savefig("plots/genetic_risk_vs_cancer_level.png", bbox_inches='tight')
plt.show()
```

### Visualize Chronic Lung Disease vs. Cancer Level

```python
# Visualize Chronic Lung Disease vs. Cancer Level
plt.figure(figsize=(12, 7))
sns.violinplot(data=df, x='Level', y='chronic Lung Disease', order=['Low', 'Medium', 'High'], palette='magma')
plt.title('Distribution of Chronic Lung Disease by Cancer Level', fontsize=16)
plt.xlabel('Cancer Level', fontsize=12)
plt.ylabel('Chronic Lung Disease Intensity', fontsize=12)
plt.savefig("plots/chronic_lung_disease_vs_cancer_level.png", bbox_inches='tight')
plt.show()
```

### HEATMAP

```python
from sklearn.preprocessing import LabelEncoder

# Create a copy of the original DataFrame to avoid modifying the original data
df_corr = df.copy()

# Use LabelEncoder to convert all object-type columns to numeric values
for col in df_corr.select_dtypes(include='object').columns:
    df_corr[col] = LabelEncoder().fit_transform(df_corr[col])

# Compute the correlation matrix
corr_matrix = df_corr.corr()

# Plot the heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of All Features', fontsize=20)
plt.savefig("plots/correlation_matrix.png", bbox_inches='tight')
plt.show()
```
